```python
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, MinMaxScaler, StandardScaler, RobustScaler
import warnings
warnings.filterwarnings("ignore")
```

```python
data = sns.load_dataset('tips')  # On récupère ici le dataset 'tips' de seaborn
data.head(3)
data.describe()
```

```python
plt.figure(figsize=(6, 6));
pd.DataFrame(data.groupby("time")["tip"].sum()).plot(kind="bar", rot = 0, xlabel = " Time", ylabel = "Somme totale");
```

```python
data[data["tip"] == data["tip"].max()]
#data.drop(170, inplace = True)
data[data["time"] == "Lunch"].tail()
```

```python
plt.figure(figsize = (10,7))
plt.scatter(data["total_bill"], data["tip"])
plt.title("Tips par rapport à l'addition")
plt.xlabel('Additions')
plt.ylabel('Tips')
sns.regplot(data = data, x = "total_bill", y = "tip");
```

```python
y = data["tip"]
X = data.drop(columns=["tip"])
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 1957)
```

```python
def preprocessing_num(train, test, model):
    columns_name = X_train.select_dtypes(include = ["float64", "int64"]).columns.tolist()
    model.fit(train.select_dtypes(include = ["float64", "int64"]))
    scaled_X_train_num = pd.DataFrame(model.transform(train.select_dtypes(include = ["float64", "int64"])))
    scaled_X_test_num = pd.DataFrame(model.transform(test.select_dtypes(include = ["float64", "int64"])))
    scaled_X_train_num = scaled_X_train_num.rename(columns = {i:f"{columns_name[i]}" for i in range(len(columns_name))})
    scaled_X_test_num = scaled_X_test_num.rename(columns = {i:f"{columns_name[i]}" for i in range(len(columns_name))})
    return scaled_X_train_num, scaled_X_test_num

           

def preprocessing_cat(train,test, model):
    model.fit(train.select_dtypes(exclude = ["float64", "int64"]))
    scaled_X_train_cat = pd.DataFrame.sparse.from_spmatrix(model.transform(train.select_dtypes(exclude = ["float64", "int64"])))
    scaled_X_test_cat = pd.DataFrame.sparse.from_spmatrix(model.transform(test.select_dtypes(exclude = ["float64", "int64"])))
    return scaled_X_train_cat, scaled_X_test_cat



def scaler_train(train, test, cat_model, num_model): 
    return pd.concat([preprocessing_cat(train,test, cat_model)[0], preprocessing_num(train, test, num_model)[0]], axis=1)

def scaler_test(train, test, cat_model, num_model): 
    return pd.concat([preprocessing_cat(train,test, cat_model)[1], preprocessing_num(train, test, num_model)[1]], axis=1)




scaler_train(X_train, X_test, OneHotEncoder(), StandardScaler()).head(10)
scaler_test(X_train, X_test, OneHotEncoder(), StandardScaler()).head(3)
```

```python
bill_scaled_train = pd.DataFrame(scaler_train(X_train, X_test, OneHotEncoder(), RobustScaler())[["total_bill", "size"]])
#bill_scaled_train.head()
bill_scaled_test = pd.DataFrame(scaler_test(X_train, X_test, OneHotEncoder(), RobustScaler())[["total_bill", "size"]])
#bill_scaled_test.head()
```

```python
lr = LinearRegression()
```

```python
lr.fit(bill_scaled_train, y_train)
lr.score(bill_scaled_test, y_test)
```

```python
y_pred = pd.DataFrame(lr.predict(bill_scaled_test))
y_pred
real_tip = pd.DataFrame(y_test.reset_index()).drop(columns="index")
my_tip = pd.concat([y_pred, real_tip, X_test[["total_bill", "size"]].reset_index().drop(columns = 'index')], axis = 1).rename(columns={0:"y_pred"})
#my_tip
```

```python
plt.figure(figsize = (10, 7))
sns.scatterplot(data = my_tip,  x = "total_bill", y = "y_pred", label = "Prédictions", s=np.exp(my_tip["size"]))
sns.scatterplot(data = my_tip,  x = "total_bill", y = "tip", label = "Real Tip", s=np.exp(my_tip["size"]))
plt.xlabel('Total bill', fontsize=20)
plt.legend()
plt.show()
```

```python
sns.scatterplot(data = my_tip, x = "tip", y = my_tip["tip"] - my_tip["y_pred"])
sns.lineplot(data = my_tip, x = my_tip["tip"], y = my_tip["y_pred"] *0);
```

```python
sns.displot(my_tip["tip"] - my_tip["y_pred"], kde = True)
plt.xlabel('Error')
plt.show();
```

```python
from sklearn.model_selection import cross_validate
cross_validate(lr, bill_scaled_train, y_train, cv=5)
cross_validate(lr, bill_scaled_train, y_train, cv=5)["test_score"].mean()
```
