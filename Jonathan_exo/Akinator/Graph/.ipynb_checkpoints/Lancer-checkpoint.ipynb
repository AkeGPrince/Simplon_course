```python
import pandas as pd 
import matplotlib.pyplot as plt

df = pd.read_csv("Akinator_.csv")
pd.set_option("display.max.columns", None)

df.drop(["Unnamed: 0"], axis = 1, inplace = True)
df.head()
```

```python
#df = df.applymap(lambda x: str(x) if type(x) == float or int else x)
df = df.applymap(str)
df = df.applymap(lambda x: x.replace(",", "."))
df = df.astype(float)
#OR 
#df = df.applymap(lambda x: float(x) if type(x) == str else x)
```

```python
x = df["Groupe 1"].to_list()
x2 = df["Groupe 2"].to_list()
x3 = df["Groupe 3"].to_list()
x4 = df["Groupe 4"].to_list()
x5 = df["Groupe 5"].to_list()
x6 = df["Groupe 6"].to_list()
x7 = df["Groupe 7"].to_list()
```

```python
fig, ax = plt.subplots()

plt.hist(x)
plt.show()
plt.plot( x, '-', color ='black')
```

```python
plt.figure(figsize = (30,10))
for i in range(df.shape[1]):
    plt.subplot(3,3, i + 1)
    plt.hist(df[f"Groupe {i + 1}"])
    plt.title(f"Groupe {i + 1}")
plt.show()
```
