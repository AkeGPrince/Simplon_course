# Exercice optionnel sur le scraping:
"Scraping en cascade"
- Scraper la page suivante: https://www.charlesbordet.com/fr/blog/
- Récupérer la liste des articles
- Pour chaque articles retrouver l'url vers la page dédiée à cet article (par exemple pour le 1er article l'adresse est : https://www.charlesbordet.com/fr/reactive-shiny/)
- Une fois arrivé sur cette page récupérer le sommaire
- Produire une liste d'articles comprenant 
    - Le titre
    - La description
    - Le tps de lecture
    - (Et surtout) le sommaire 


```python
import requests as req
from bs4 import BeautifulSoup
import pandas as pd 
articles_url = f"https://www.charlesbordet.com/"
url = f"https://www.charlesbordet.com/fr/blog/#"
page = req.get(url)
soup = BeautifulSoup(page.content, "html.parser")
zone = soup.find(class_="archive")
zone2 = soup.find(class_="toc__menu")
articles = []
titles = []
descriptions = []
reading_time = []
sommaire = []
```

```python
for i in range(len(zone.find_all("a"))):
    balise = zone.find_all("a")[i]
    #print(balise)
    link = balise['href']
    articles_url = f"https://www.charlesbordet.com/" + link
    balise_title, balise_time, balise_decritpions = zone.find_all("a")[i], zone.find_all(class_="page__meta")[i], zone.find_all(class_="archive__item-excerpt")[i]
    articles_name, read_time, resume = balise_title.text, balise_time.text, balise_decritpions.text
    if articles_url not in articles:
        articles.append(articles_url)
        continue
    elif articles_name not in titles and read_time not in reading_time and resume not in descriptions:
        titles.append(articles_name)
        reading_time.append(read_time)
        descriptions.append(resume)
        continue
        for x in range(len(articles)):
            page = req.get(articles[x])
            soup = BeautifulSoup(page.content, "html.parser")
            zone2 = soup.find(class_="toc__menu")
            for i in range(len(zone2)):
                somm = zone2.find_all("a")
                #somm1 = somm[i].text
                #print(somm1)
                #print(len(somm))
                for i in range(len(somm)):
                    if somm[i].text not in sommaire:
                        #print(somm[i].text)
                        sommaire.append(somm[i].text)
    
```
