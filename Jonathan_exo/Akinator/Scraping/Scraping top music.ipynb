```python
import pandas as pd 
import requests as req
from bs4 import BeautifulSoup


data = pd.read_csv("akinator.csv")
birth_date = data["Anniversaire"]
birth_clean = birth_date.fillna(0)
name = data["Prénom"].tolist()
surname = data["Nom"].tolist()
full_name = list(zip(surname, name))
full_name
dates=[]
artiste_song = []
years = {}

for date in birth_clean:
    try:
        x = date.split('/')
        new_date = "/".join((x[2], x[1], x[0]))
        dates.append(new_date)
    except:
        dates.append("0")
#dates
for i in range(len(dates)):
    try:
        url = req.get(f"https://www.birthdayjams.com/us/{dates[i]}/")
        soup = BeautifulSoup(url.content, "html.parser")
        artiste = soup.find("div", class_="prelims hidden-xs").get_text(" | ", strip=True) 
        artiste_song.append(artiste)
    except:
        artiste_song.append("No artist found")
#artiste_song

dates_song = list(zip(dates, artiste_song))
#dates_song

for i in range(len(full_name)):
    years[full_name[i]] = dates_song[i]
print(years)

tab = pd.DataFrame(years)
tab
```
