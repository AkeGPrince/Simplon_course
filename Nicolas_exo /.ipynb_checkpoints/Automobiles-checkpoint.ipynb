#### Partie 1

```python
import pandas as pd 

df = pd.read_csv('Automobiles.csv')

df.drop(["nbSeats", 'wheel-base-style-circum-diam', 'average-mileage'], inplace= True, axis=1)
df.rename(columns={"num-of-cylinders": "cylinders"}, inplace = True)
```

```python
df['cylinders'] = df['cylinders'].replace(['two', 'twelve', 'five5', 'six6', '[four4', 'three', 'eight'],
                                          [2, 12, 5, 6, 4, 3, 8])
```

```python
df['color'] = df['color'].apply(lambda x: x.replace('color-', '') if str(x).startswith('color-') else x)
df['color'] = df['color'].apply(lambda x: 'black' if 'blac' in str(x) else x )
#df['color'] = df['color'].apply(lambda x: x.replace('blac', 'black') if str(x).startswith('blac') else x)
df['length'] = df['length'].round(2)
```

```python
df1 = df.copy()
```

```python
df1['color'].fillna(method='bfill', inplace = True)
```

```python
df1.loc[df1['price'].idxmax()]
```

```python
df1.count() / df1.shape[0] * 100
```
